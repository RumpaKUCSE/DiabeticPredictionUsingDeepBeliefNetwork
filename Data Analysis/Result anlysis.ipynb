```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
xls = pd.ExcelFile("Result analysis.xlsx")
print(xls.sheet_names)
dfstd = pd.read_excel('Result analysis.xlsx', sheetname ='10-8-0.01-5000' )

dfl5 = pd.read_excel('Result analysis.xlsx', sheetname ='10-5-0.01-5000' )
dfl15 = pd.read_excel('Result analysis.xlsx', sheetname ='10-15-0.01-5000' )

dfu15 = pd.read_excel('Result analysis.xlsx', sheetname ='15-8-0.01-5000' )
dfu7 = pd.read_excel('Result analysis.xlsx', sheetname ='7-8-0.01-5000' )

dfr7 = pd.read_excel('Result analysis.xlsx', sheetname ='10-8-0.7-5000' )
dfr001 = pd.read_excel('Result analysis.xlsx', sheetname ='10-8-0.001-5000' )

dfi4500 = pd.read_excel('Result analysis.xlsx', sheetname ='10-8-0.01-4500' )
dfi5500 = pd.read_excel('Result analysis.xlsx', sheetname ='10-8-0.01-5500' )


epochs = dfstd.loc[:,['no-epoch']].values
npep = np.array(epochs)

std = dfstd.loc[:,['log-error']].values
npstd = np.array(std)

l5 = dfl5.loc[:,['log-error']].values
npl5 = np.array(l5)
l15 = dfl15.loc[:,['log-error']].values
npl15 = np.array(l15)

u15 = dfu15.loc[:,['log-error']].values
npu15 = np.array(u15)
u7 = dfu7.loc[:,['log-error']].values
npu7 = np.array(u7)

r7 = dfr7.loc[:,['log-error']].values
npr7 = np.array(r7)
r001 = dfr001.loc[:,['log-error']].values
npr001 = np.array(r001)

i4500 = dfi4500.loc[:,['log-error']].values
npi4500 = np.array(i4500)
i5500 = dfi5500.loc[:,['log-error']].values
npi5500 = np.array(i5500)


```

```python
plt.plot(npep, npstd, linestyle = '-', color = 'blue', linewidth = 2)
plt.plot(npep,npl5, linestyle = ':', color = 'red', linewidth = 2)
plt.plot(npep,npl15, linestyle = ':', color = 'green', linewidth = 2)
plt.axis([1,22,.4,.7])
plt.xticks(np.arange(0,21,step=1))
plt.xlabel('no. of epochs')
plt.ylabel('ANN training loss')

plt.text(17.7, .625, '______ 8 \n\n',color='blue',fontsize=12,
        bbox={'facecolor':'white', 'alpha':0.5, 'pad':10})

plt.text(17.7, .6, '.......... 15 \n\n',color='red',fontsize=12)

plt.text(17.7, .58, '.......... 5 \n\n',color='green',fontsize=12)

plt.title('Different Number of Layer')


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 4
men_means = (83.12,77.08, 89.22, 71.15)


ind = np.arange(N)  # the x locations for the groups
width = 0.6      # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind, men_means, width, color='b')


# add some text for labels, title and axes ticks
ax.set_ylabel('Scores in percentage')

#ax.set_xticks(ind + width / 2)
plt.xticks(ind,('Accuracy', 'Precision', 'Sensitivity', 'Specificity'))
plt.yticks(np.arange(0,110,step=10))

#ax.legend((rects1[0], rects2[0]), ('Men', 'Women'))


def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.042*height,
                '%.2f' % float(height),
                ha='center', va='bottom')

autolabel(rects1)


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 4
men_means = (83.12,89.22,71.15, (100-83.12))
men_std = ( 3, 4, 1, 2)

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind, men_means, width, color='b', yerr=men_std)

women_means = (80.72, 88.4,64.5,19.27)
women_std = ( 5, 2, 3, 3)
rects2 = ax.bar(ind + width, women_means, width, color='g', yerr=women_std)

# add some text for labels, title and axes ticks
ax.set_ylabel('Scores in percentage')
#ax.set_title('Scores by group and gender')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('Accuracy', 'Sensitivity', 'Specificity', 'Error Rate'))
plt.yticks(np.arange(0,110,step=10))
ax.legend((rects1[0], rects2[0]), ('Proposed Method', 'Vijayan et al.'))


def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%.2f' % float(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.show()
```
