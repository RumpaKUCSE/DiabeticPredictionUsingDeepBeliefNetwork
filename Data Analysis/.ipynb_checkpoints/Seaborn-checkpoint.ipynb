```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("diabetes.csv")
df[0:7]
```

```python
sns.lmplot(x='BloodPressure', y='DiabetesPedigreeFunction', data=df,
           fit_reg=False, hue = 'Outcome' # No regression line
           )

plt.ylim(0, None)
plt.xlim(0, None)
```

```python
sns.boxplot(data = df)
```

```python
# Pre-format DataFrame
stats_df = df.drop(['Outcome'], axis=1)
 
# New boxplot using stats_df
sns.boxplot(data=stats_df)
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(30,30))
# Set theme
sns.set_style('darkgrid')
 
# Violin plot
sns.violinplot(x='BloodPressure', y='DiabetesPedigreeFunction', data=df)
```

```python
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
```

```python
# Violin plot with Pokemon color palette
plt.figure(figsize=(30,30))
sns.violinplot(x='BloodPressure', y='DiabetesPedigreeFunction', data=df, 
               palette=pkmn_type_colors) # Set color palette
```

```python
plt.figure(figsize=(30,30))
sns.swarmplot(x='BloodPressure', y='DiabetesPedigreeFunction', data=df, 
              palette=pkmn_type_colors)
```

```python
plt.figure(figsize=(30,30))

sns.violinplot(x='BloodPressure', y = 'DiabetesPedigreeFunction',data = df, inner = None, palette=pkmn_type_colors)

sns.swarmplot(x='BloodPressure', 
              y='DiabetesPedigreeFunction', 
              data=df, 
              color='k', # Make points black
              alpha=0.7) # and slightly transparent
 
# Set title with matplotlib
plt.title('DiastolicBloodPressure vs DiabetesPedigree')
```

```python
# Melt DataFrame
melted_df = pd.melt(stats_df, 
                    id_vars=["Pregnancies", "BloodPressure", "Age"], # Variables to keep
                    var_name="Stat") # Name of melted variable
melted_df.head()
```

```python
print( stats_df.shape )
print( melted_df.shape )
```

```python
sns.swarmplot(x='Stat', y='value', data=melted_df, 
              hue='Pregnancies')
```

```python
# 1. Enlarge the plot
plt.figure(figsize=(10,6))
 
sns.swarmplot(x='Stat', 
              y='value', 
              data=melted_df, 
              hue='Pregnancies', 
              dodge=True, # 2. Separate points by hue
              palette=pkmn_type_colors) # 3. Use Pokemon palette
 
# 4. Adjust the y-axis
plt.ylim(0, 800)
 
# 5. Place legend to the right
plt.legend(bbox_to_anchor=(1, 1), loc=2)
```

```python
# Calculate correlations
corr = df.corr()
 
# Heatmap
sns.heatmap(corr)  
```

```python
sns.distplot(df.BMI)
```

```python
# Count Plot (a.k.a. Bar Plot)
sns.countplot(x='Pregnancies', data=df, palette=pkmn_type_colors)
 
# Rotate x-labels
plt.xticks(rotation=45)
```

```python
# Factor Plot
g = sns.factorplot(x='Age', 
                   y='Glucose', 
                   data=df, 
                   hue='Pregnancies',  # Color by stage
                   col='Pregnancies',  # Separate by stage
                   kind='swarm') # Swarmplot
 
# Rotate x-axis labels
g.set_xticklabels(rotation=90)
 
# Doesn't work because only rotates last plot
# plt.xticks(rotation=-45)
plt.figure(figsize=(30,30))
```

```python
sns.kdeplot(df.Age, df.SkinThickness)
```

```python
sns.jointplot(x='Age', y='SkinThickness', data=df)
```
